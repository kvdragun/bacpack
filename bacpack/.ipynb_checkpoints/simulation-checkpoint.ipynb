```python
#@markdown Importing packages
import sys
import os
from ipywidgets import FloatProgress,Text,Checkbox, HBox, Image
from IPython.display import display, HTML
import numpy as np
import pandas as pd
import math
from numba import cuda
from cycler import cycler
import matplotlib.pyplot as plt
import matplotlib
import re
from bac import Sim_BN
from highfreq import TSRV_imp
```

```python
#@markdown functions definitions
#sys.path.append('gdrive/MyDrive/BAC_code')

def sim_gpu(assets=10,trades=270000,low=0.9,high=0.99,runs=np.arange(10,dtype=np.int32),noise_kappa=0,
            K=2,jumpcont=False,jumpsperiod=0,jumpmagnitude=1,jeta=25, savep=None, 
            factor=5,minf=0.01,skip=False):
    f = FloatProgress(min=np.min(runs), max=np.max(runs))
    display(f)
    wt_noise=Text(
      value='',
      description='ETF noise:',
      disabled=True)
    display(wt_noise)
    
    wt_est=Text(
      value='',
      description='Pre-estimator:',
      disabled=True)
    display(wt_est)

    resR=pd.DataFrame(np.zeros((12,3)),index=
    [
     "EST",
     "NEST",                     
     "NBAC_HY_NR_Wk",
     "NBAC_oracle_NR_Wk",
     "NBAC_02_NR_Wk",
     "NBAC_05_NR_Wk",
     "NBAC_VAB_NR_Wk",
     
     "BAC_HY_NR_Wk",
     "BAC_oracle_NR_Wk",
     "BAC_02_NR_Wk",
     "BAC_05_NR_Wk",
     "BAC_VAB_NR_Wk"
     
     ], columns=['HY','RC','TSC'])
    csim=Sim_BN(assets,trades,low,high)
    csim.frequencies(factor,minf)
    for l in runs:
        f.value=l
        if savep!=None and skip:
            for j in resR.columns:
                nj="_noise%.2f"%noise_kappa if noise_kappa>0 else ""
                nj+="_jump" if jumpcont else ""
                tmp=savep+"/%d"%assets+j+nj+"/_%d"%l+".npz"
                if not os.path.exists(tmp):
                    #print(tmp)
                    break
            else:
              continue
        csim.generate(True,fexp=False,
                      jumpsperiod=jumpsperiod,jumpmagnitude=jumpmagnitude)
        etf_nf=csim.etf.copy()
        if noise_kappa>0:
            csim.noise(noise_kappa)
            mm0_01=csim.fvar()
            nf_diag_01=csim.TSRV(2)
            noise_vr=(mm0_01-nf_diag_01)/2/csim.missing_points.sum(axis=0)
            noise_vr=np.maximum(noise_vr,0)
            etfvpn=(csim.letf[csim.mp_etf]**2).sum()
            etf_noise2=np.maximum(etfvpn-TSRV_imp(csim.letf,csim.mp_etf,2,1),0)
        else:
            noise_vr=np.zeros(csim.numvar)
            nf_diag_01=csim.fvar()
            etfvpn=(csim.letf[csim.mp_etf]**2).sum()
            etf_noise2=0

        mt=np.dot(csim.comps.T,csim.comps)
        
        if jumpcont:
            adj=csim.numper/(csim.numper-csim.jumpsfree(nf_diag_01,noise_vr,
                              etfvpn,etf_noise2/2/csim.mp_etf.sum()))
            mm0_01=csim.fvar()
            nf_diag_01=csim.TSRV(2)
            noise_vr=(mm0_01-nf_diag_01)/2/csim.missing_points.sum(axis=0)
            noise_vr=np.maximum(noise_vr,0)
            if noise_kappa>0:
                mm0_01=csim.fvar()
                nf_diag_01=csim.TSRV(2)
                noise_vr=(mm0_01-nf_diag_01)/2/csim.missing_points.sum(axis=0)
                noise_vr=np.maximum(noise_vr,0)
                etfvpn=(csim.letf[csim.mp_etf]**2).sum()
                etf_noise2=np.maximum(etfvpn-TSRV_imp(csim.letf,csim.mp_etf,2,1),0)
            else:
                noise_vr=np.zeros(csim.numvar)
                nf_diag_01=csim.fvar()
                etfvpn=(csim.letf[csim.mp_etf]**2).sum()
                etf_noise2=0
        else:
            adj=1
            
        wt_noise.value="%f"%etf_noise2
        
        for j in resR.columns:
            wt_est.value=j
            c_m0_HY=np.zeros((csim.numvar,csim.numvar))
            c_m0_NHY=np.zeros((csim.numvar,csim.numvar))
                        
            c_m0_NBAC_HY_HY_NR_Wk=np.zeros((csim.numvar,csim.numvar))
            c_m0_NBAC_HY_oracle_NR_Wk=np.zeros((csim.numvar,csim.numvar))
            c_m0_NBAC_HY_02_NR_Wk=np.zeros((csim.numvar,csim.numvar))
            c_m0_NBAC_HY_05_NR_Wk=np.zeros((csim.numvar,csim.numvar))
            c_m0_NBAC_HY_VAB_NR_Wk=np.zeros((csim.numvar,csim.numvar))
            
            c_m0_BAC_HY_HY_NR_Wk=np.zeros((csim.numvar,csim.numvar))
            c_m0_BAC_HY_oracle_NR_Wk=np.zeros((csim.numvar,csim.numvar))
            c_m0_BAC_HY_02_NR_Wk=np.zeros((csim.numvar,csim.numvar))
            c_m0_BAC_HY_05_NR_Wk=np.zeros((csim.numvar,csim.numvar))
            c_m0_BAC_HY_VAB_NR_Wk=np.zeros((csim.numvar,csim.numvar))
            
            b_low=0
            e_up=trades

            if True:
                na_diag=2*noise_vr*csim.missing_points[b_low:e_up,:].sum(axis=0)
                mweights=csim.meanw(b_low=b_low,e_up=e_up)
                
                if j=="RC":
                    m0_HY,b0_HY_W=csim.RCov()
                    
                if j=="HY":
                    m0_HY,b0_HY_W=csim.HY()
                
                if j=="TSC":
                    tsc=csim.TSCov(K=K)
                    m0_HY=tsc
                    b0_HY_W=np.dot(tsc,mweights)
                    na_diag=np.zeros_like(na_diag)
                
                mtnorm=np.linalg.norm(mt)
                b_RC=csim.beta(b_low=b_low,e_up=e_up)                    
                b_oracle=np.dot(csim.comps.T,etf_nf)
                if noise_kappa>0:
                    part_etf_noise=(e_up-b_low)/csim.numper*etf_noise2
                    
                    na_HY=na_diag
                    b0N_HY_Wk=b0_HY_W-na_HY*mweights
                    
                    bN_RC=b_RC#csim.dgpu_TSCov_beta(K=50,J=10)
                    bN_02=0.2*(bN_RC-b_oracle)+b_oracle
                    bN_05=0.5*(bN_RC-b_oracle)+b_oracle
                    bN_VAB=b_RC+csim.fBAC_beta_log(etf_noise=part_etf_noise)
                    
                    c_m0_NHY+=m0_HY-np.diag(na_HY)
                    
                    tmp_Delta,NL_NR=csim.NBAC_Delta_NR(b0N_HY_Wk,bN_RC,na_HY,b_low=b_low,
                                                     e_up=e_up,L_out=True)
                    c_m0_NBAC_HY_HY_NR_Wk+=m0_HY-tmp_Delta
                    c_m0_NBAC_HY_oracle_NR_Wk+=m0_HY-csim.NBAC_Delta_NR(b0N_HY_Wk,
                                                b_oracle,na_HY,L_in=True,L=NL_NR)
                    c_m0_NBAC_HY_02_NR_Wk+=m0_HY-csim.NBAC_Delta_NR(b0N_HY_Wk,bN_02,
                                                        na_HY,L_in=True,L=NL_NR)
                    c_m0_NBAC_HY_05_NR_Wk+=m0_HY-csim.NBAC_Delta_NR(b0N_HY_Wk,bN_05,
                                                        na_HY,L_in=True,L=NL_NR)
                    c_m0_NBAC_HY_VAB_NR_Wk+=m0_HY-csim.NBAC_Delta_NR(b0N_HY_Wk,
                                                bN_VAB,na_HY,L_in=True,L=NL_NR)
                    c_m0_NHY*=adj
                    c_m0_NBAC_HY_HY_NR_Wk*=adj
                    c_m0_NBAC_HY_oracle_NR_Wk*=adj
                    c_m0_NBAC_HY_02_NR_Wk*=adj
                    c_m0_NBAC_HY_05_NR_Wk*=adj
                    c_m0_NBAC_HY_VAB_NR_Wk*=adj

                    resR.loc["NEST",j]+=np.linalg.norm(mt-c_m0_NHY)**2/mtnorm**2
                    resR.loc["NBAC_HY_NR_Wk",j]+=np.linalg.norm(mt-c_m0_NBAC_HY_HY_NR_Wk)**2/mtnorm**2
                    resR.loc["NBAC_oracle_NR_Wk",j]+=np.linalg.norm(mt-c_m0_NBAC_HY_oracle_NR_Wk)**2/mtnorm**2
                    resR.loc["NBAC_02_NR_Wk",j]+=np.linalg.norm(mt-c_m0_NBAC_HY_02_NR_Wk)**2/mtnorm**2
                    resR.loc["NBAC_05_NR_Wk",j]+=np.linalg.norm(mt-c_m0_NBAC_HY_05_NR_Wk)**2/mtnorm**2
                    resR.loc["NBAC_VAB_NR_Wk",j]+=np.linalg.norm(mt-c_m0_NBAC_HY_VAB_NR_Wk)**2/mtnorm**2
                    
                else:
                    b0_HY_Wk=b0_HY_W                
                    b_02=0.2*(b_RC-b_oracle)+b_oracle
                    b_05=0.5*(b_RC-b_oracle)+b_oracle
                    b_VAB=b_RC+csim.fBAC_beta_log()
                    c_m0_HY+=m0_HY
                    
                    tmp_Delta,L_NR=csim.BAC_Delta_NR(b0_HY_Wk,b_RC,b_low=b_low,
                                                     e_up=e_up,L_out=True)
                    c_m0_BAC_HY_HY_NR_Wk+=m0_HY-tmp_Delta
                    
                    c_m0_BAC_HY_oracle_NR_Wk+=m0_HY-csim.BAC_Delta_NR(b0_HY_Wk,
                                                    b_oracle,L_in=True,L=L_NR)
                    c_m0_BAC_HY_02_NR_Wk+=m0_HY-csim.BAC_Delta_NR(b0_HY_Wk,b_02,
                                                        L_in=True,L=L_NR)
                    c_m0_BAC_HY_05_NR_Wk+=m0_HY-csim.BAC_Delta_NR(b0_HY_Wk,b_05,
                                                            L_in=True,L=L_NR)
                    c_m0_BAC_HY_VAB_NR_Wk+=m0_HY-csim.BAC_Delta_NR(b0_HY_Wk,b_VAB,
                                                            L_in=True,L=L_NR)
                    c_m0_HY*=adj
                    c_m0_BAC_HY_HY_NR_Wk*=adj
                    c_m0_BAC_HY_oracle_NR_Wk*=adj
                    c_m0_BAC_HY_02_NR_Wk*=adj
                    c_m0_BAC_HY_05_NR_Wk*=adj
                    c_m0_BAC_HY_VAB_NR_Wk*=adj
                    resR.loc["EST",j]+=np.linalg.norm(mt-c_m0_HY)**2/mtnorm**2
                    resR.loc["BAC_HY_NR_Wk",j]+=np.linalg.norm(mt-c_m0_BAC_HY_HY_NR_Wk)**2/mtnorm**2
                    resR.loc["BAC_oracle_NR_Wk",j]+=np.linalg.norm(mt-c_m0_BAC_HY_oracle_NR_Wk)**2/mtnorm**2
                    resR.loc["BAC_02_NR_Wk",j]+=np.linalg.norm(mt-c_m0_BAC_HY_02_NR_Wk)**2/mtnorm**2
                    resR.loc["BAC_05_NR_Wk",j]+=np.linalg.norm(mt-c_m0_BAC_HY_05_NR_Wk)**2/mtnorm**2
                    resR.loc["BAC_VAB_NR_Wk",j]+=np.linalg.norm(mt-c_m0_BAC_HY_VAB_NR_Wk)**2/mtnorm**2                
            
            if savep!=None:
                nj="_noise%.2f"%noise_kappa if noise_kappa>0 else ""
                nj+="_jump" if jumpcont else ""
                if not os.path.exists(savep+"/%d"%assets+j+nj):
                    os.mkdir(savep+"/%d"%assets+j+nj)
                    
                if noise_kappa>0:    
                    np.savez_compressed(savep+"/%d"%assets+j+nj+"/_%d"%l,
                    preest=c_m0_HY,
                    npreest=c_m0_NHY,
                    NBAC_HY=c_m0_NBAC_HY_HY_NR_Wk,
                    NBAC_oracle=c_m0_NBAC_HY_oracle_NR_Wk,
                    NBAC_02=c_m0_NBAC_HY_02_NR_Wk,
                    NBAC_05=c_m0_NBAC_HY_05_NR_Wk,
                    NBAC_VAB=c_m0_NBAC_HY_VAB_NR_Wk)
                else:   
                    np.savez_compressed(savep+"/%d"%assets+j+nj+"/_%d"%l,
                    preest=c_m0_HY,
                    npreest=c_m0_NHY,
                    BAC_HY=c_m0_BAC_HY_HY_NR_Wk,
                    BAC_oracle=c_m0_BAC_HY_oracle_NR_Wk,
                    BAC_02=c_m0_BAC_HY_02_NR_Wk,
                    BAC_05=c_m0_BAC_HY_05_NR_Wk,
                    BAC_VAB=c_m0_BAC_HY_VAB_NR_Wk)
                
        if savep!=None:
            rcx=RCx(csim.acomps,100)
            if not os.path.exists(savep+"/%d"%assets+"TRUE"+nj):
                    os.mkdir(savep+"/%d"%assets+"TRUE"+nj)
            np.savez_compressed(savep+"/%d"%assets+"TRUE"+nj+"/_%d"%l,
                true=csim.ssCov,
                highprec=mt,
                RCx=rcx)
            (resR/len(runs)).to_excel(savep+"/res"+"%d"%assets+nj+".xlsx")    
    #return resR/len(runs)

    
def RCx(acomps,size):
    ac=acomps.reshape(acomps.shape[0]//size,size,acomps.shape[1]).sum(axis=1)
    return np.dot(ac.T,ac)

def sim_gpu_noise(assets=10,trades=270000,low=0.9,high=0.99,runs=np.arange(10),
                  noise_kappa=np.arange(0.1),K=30, jumpcont=False,jumpsperiod=0,
                  jumpmagnitude=1, jeta=25, savep=None, 
                  factor=10,minf=0.02, skip=False):
    resR=pd.DataFrame(np.zeros((len(noise_kappa),13)),index=
    noise_kappa, columns=['HY','RC','TSC',"HYJ","RCJ","TSCJ",
                    'oHY','oRC','oTSC',"oHYJ","oRCJ","oTSCJ","adj"])
    csim=Sim_BN(assets,trades,low,high)
    csim.frequencies(factor,minf)
    f = FloatProgress(min=np.min(runs), max=np.max(runs))
    f2 = FloatProgress(min=np.min(noise_kappa), max=np.max(noise_kappa))
    display(f)
    display(f2)
    wt_kappa=Text(
      value='',
      description='kappa:',
      disabled=True)
    #display(wt_kappa)
    wt_est=Text(
      value='',
      description='Pre-estimator:',
      disabled=True)
    #display(wt_est)
    wt_noise=Text(
      value='',
      description='ETF noise:',
      disabled=True)
    #display(wt_noise)
    hb=HBox([wt_kappa,wt_est,wt_noise])
    display(hb)
    for l in runs:
        f.value=l
        if skip:
         if not jumpcont:
          jmp=0 
          for jns in resR.index:
            nj="_noise%.3f"%jns 
            nj+="_jump" if jumpcont and jmp==1 else ""
            tmp=savep+"/%d"%assets+"TRUE"+nj+"/_%d"%l+".npz"
            if not os.path.exists(tmp):
              print(tmp)
              break
          else:    
            continue
         else:
          jmp=1
          for jns in resR.index:
            nj="_noise%.3f"%jns 
            nj+="_jump" if jumpcont and jmp==1 else ""
            tmp=savep+"/%d"%assets+"TRUE"+nj+"/_%d"%l+".npz"
            if not os.path.exists(tmp):
              print(tmp)
              break
          else:    
            continue   
        csim.generate(True,fexp=False)
        for jmp in range(2 if jumpcont>0 else 1) :
            if jmp==1:
                csim.gen_jmp(jumpsperiod=jumpsperiod,jumpmagnitude=jumpmagnitude)
            sacomps=csim.acomps
            sletf=csim.letf
            setf=csim.etf
            for jns in resR.index:
                f2.value=jns
                            
                wt_kappa.value="%f"%jns
                csim.acomps=sacomps.copy()
                csim.letf=sletf.copy()
                csim.etf=setf.copy()
                etf_nf=csim.etf.copy()
                if jns>0:
                    csim.noise(jns)
                    etfvpn=(csim.letf[csim.mp_etf]**2).sum()
                    etf_noise2=np.maximum(etfvpn-TSRV_imp(csim.letf,csim.mp_etf,2,1),0)
                else:
                    etfvpn=(csim.letf[csim.mp_etf]**2).sum()
                    etf_noise2=0
                mm0_01=csim.fvar()
                nf_diag_01=csim.TSRV(2)
                noise_vr=(mm0_01-nf_diag_01)/2/csim.missing_points.sum(axis=0)
                noise_vr=np.maximum(noise_vr,0)
                mt=np.dot(csim.comps.T,csim.comps)
                
                if jumpcont and jmp==1:
                    adj=csim.numper/(csim.numper-csim.jumpsfree(nf_diag_01,noise_vr,
                                     etfvpn,etf_noise2/2/csim.mp_etf.sum()))
                    mm0_01=csim.fvar()
                    nf_diag_01=csim.TSRV(2)
                    noise_vr=(mm0_01-nf_diag_01)/2/csim.missing_points.sum(axis=0)
                    noise_vr=np.maximum(noise_vr,0)
                    etfvpn=(csim.letf[csim.mp_etf]**2).sum()
                    etf_noise2=np.maximum(etfvpn-TSRV_imp(csim.letf,csim.mp_etf,2,1),0)  
                else:
                    adj=1
                    #etfvpn=(csim.letf[csim.mp_etf]**2).sum()
                    #etf_noise2=0
                wt_noise.value="%f"%etf_noise2
                for j in range(3):
                    sest=["HY","RC","TSC"][j]
                    wt_est.value=sest
                    c_m0_HY=np.zeros((csim.numvar,csim.numvar))
                    c_m0_NHY=np.zeros((csim.numvar,csim.numvar))
                                
                    c_m0_NBAC_HY_HY_NR_Wk=np.zeros((csim.numvar,csim.numvar))
                    c_m0_NBAC_HY_oracle_NR_Wk=np.zeros((csim.numvar,csim.numvar))
                    c_m0_NBAC_HY_VAB_NR_Wk=np.zeros((csim.numvar,csim.numvar))
                    
                    
                    b_low=0
                    e_up=trades
        
                    if True:
                        na_diag=2*noise_vr*csim.missing_points[b_low:e_up,:].sum(axis=0)
                        mweights=csim.meanw(b_low=b_low,e_up=e_up)
                        
                        if sest=="RC":
                            m0_HY,b0_HY_W=csim.fRC2()
                            
                        if sest=="HY":
                            m0_HY,b0_HY_W=csim.fHY2()
                        
                        if sest=="TSC":
                            tsc=csim.TSCov(K=K)
                            m0_HY=tsc
                            b0_HY_W=np.dot(tsc,mweights)
                            na_diag=np.zeros_like(na_diag)
                                
                        b_RC=csim.beta(b_low=b_low,e_up=e_up)                    
                        b_oracle=np.dot(csim.comps[csim.mp_etf,:].T,etf_nf[csim.mp_etf])
                        part_etf_noise=(e_up-b_low)/csim.numper*etf_noise2
                        
                        na_HY=na_diag
                        b0N_HY_Wk=b0_HY_W-na_HY*mweights
                        
                        bN_RC=b_RC#csim.dgpu_TSCov_beta(K=50,J=10)
                        bN_VAB=b_RC+csim.fBAC_beta_log(etf_noise=part_etf_noise)
                        
                        c_m0_NHY+=m0_HY-np.diag(na_HY)
                        
                        tmp_Delta,NL_NR=csim.NBAC_Delta_NR(b0N_HY_Wk,bN_RC,na_HY,b_low=b_low,
                                                         e_up=e_up,L_out=True)
                        c_m0_NBAC_HY_HY_NR_Wk+=m0_HY-tmp_Delta
                        c_m0_NBAC_HY_oracle_NR_Wk+=m0_HY-csim.NBAC_Delta_NR(b0N_HY_Wk,
                                                    b_oracle,na_HY,L_in=True,L=NL_NR)
                        c_m0_NBAC_HY_VAB_NR_Wk+=m0_HY-csim.NBAC_Delta_NR(b0N_HY_Wk,
                                                    bN_VAB,na_HY,L_in=True,L=NL_NR)
                        c_m0_NHY*=adj
                        c_m0_NBAC_HY_HY_NR_Wk*=adj
                        #c_m0_NBAC_HY_oracle_NR_Wk*=adj
                        c_m0_NBAC_HY_VAB_NR_Wk*=adj
                
                    if savep!=None:
                        nj="_noise%.3f"%jns 
                        nj+="_jump" if jumpcont and jmp==1 else ""
                        if not os.path.exists(savep+"/%d"%assets+sest+nj):
                            os.mkdir(savep+"/%d"%assets+sest+nj)   
                        np.savez_compressed(savep+"/%d"%assets+sest+nj+"/_%d"%l,
                            preest=c_m0_HY,
                            npreest=c_m0_NHY,
                            NBAC_HY=c_m0_NBAC_HY_HY_NR_Wk,
                            NBAC_oracle=c_m0_NBAC_HY_oracle_NR_Wk,
                            NBAC_VAB=c_m0_NBAC_HY_VAB_NR_Wk)  
                    if not os.path.exists(savep+"/%d"%assets+"TRUE"+nj):
                        os.mkdir(savep+"/%d"%assets+"TRUE"+nj)
                    np.savez_compressed(savep+"/%d"%assets+"TRUE"+nj+"/_%d"%l,
                            true=csim.ssCov,
                            highprec=mt)    
                    #print(resR.columns[j+jmp*2])
                    resR.loc[jns,resR.columns[j+jmp*3]]+=(np.linalg.norm(mt-c_m0_NHY)**2/
                        np.linalg.norm(mt-c_m0_NBAC_HY_VAB_NR_Wk)**2)
                    resR.loc[jns,resR.columns[j+jmp*3+6]]+=(np.linalg.norm(mt-c_m0_NHY)**2/
                        np.linalg.norm(mt-c_m0_NBAC_HY_oracle_NR_Wk)**2)
                    resR.loc[jns,"adj"]+=adj
    (resR/len(runs)).to_excel(savep+"/noise_sensitivity.xlsx")                
    return resR/len(runs)

def stats(bacpath="BAC_Data/Sim_res"): 
    #simulatuon results
    subs=os.listdir(bacpath)
    
    res=pd.DataFrame(columns=["assets","estimator","kappa","jumps","MSE pre-estimator","MSE BAC","MSE BAC oracle",
                              "MSE VAB BAC"])
    
    cnt=0
    f1 = FloatProgress(min=0, max=len(subs))
    f2 = FloatProgress(min=0, max=1)
    display(f1)
    display(f2)
    for sb in subs:
        f1.value+=1
        if sb.find("TRUE")>=0 or not os.path.isdir(bacpath+"/"+sb):
            continue
        #print(sb)
        cest=0
        cBAC_HY=0
        cBAC_oracle=0
        cBAC_02=0
        cBAC_05=0
        cBAC_VAB=0
        noisy=True
        if sb.find("noise")<0:
            noisy=False
        jump=True
        if sb.find("jump")<0:
            jump=False
        assets=re.findall("\d+.",sb)[0][:-1]
        ee=sb.find("_")
        estim=sb[len(assets):ee if ee>=0 else len(sb)]
        if noisy:
          kappa=re.findall("\d*\.\d+|\d+",sb[sb.find(" noise")+5:])[0]
        else:
          kappa=""                   
        files=os.listdir(bacpath+"/"+sb)
        qfiles=len(files)
        f2.value=0
        for f in files:
            f2.value+=1/qfiles
            if f[-3:]!="npz":
                qfiles-=1
                continue
            #print(bacpath+"/"+sb+"/"+f)
            with np.load(bacpath+"/"+sb+"/"+f) as arrs:
                if noisy:
                    est=arrs["npreest"]
                    BAC_HY=arrs["NBAC_HY"]
                    BAC_oracle=arrs["NBAC_oracle"]
                    BAC_02=arrs["NBAC_02"]
                    BAC_05=arrs["NBAC_05"]
                    BAC_VAB=arrs["NBAC_VAB"]
                else:
                    est=arrs["preest"]
                    BAC_HY=arrs["BAC_HY"]
                    BAC_oracle=arrs["BAC_oracle"]
                    BAC_02=arrs["BAC_02"]
                    BAC_05=arrs["BAC_05"]
                    BAC_VAB=arrs["BAC_VAB"]
            sbt=sb.replace("HY","TRUE").replace("RC","TRUE").replace("TSC","TRUE") 
            #print(bacpath+"/"+sbt+"/"+f)
            with np.load(bacpath+"/"+sbt+"/"+f) as arrs:
                        #true=arrs["true"]
                hp=arrs["highprec"]
                        #RCx=arrs["RCx"]
                
            hpnorm=np.linalg.norm(hp)**2*qfiles
            cest+=np.linalg.norm(hp-est)**2/hpnorm
            cBAC_HY+=np.linalg.norm(hp-BAC_HY)**2/hpnorm
            cBAC_oracle+=np.linalg.norm(hp-BAC_oracle)**2/hpnorm
            cBAC_02+=np.linalg.norm(hp-BAC_02)**2/hpnorm
            cBAC_05+=np.linalg.norm(hp-BAC_05)**2/hpnorm
            cBAC_VAB+=np.linalg.norm(hp-BAC_VAB)**2/hpnorm
        
        res.loc[cnt]=[assets,estim,kappa,jump,cest,cBAC_HY,cBAC_oracle,cBAC_VAB]
        
        #print(qfiles," files processed")
        cnt+=1
    return res

def noise_magnit_pre(bacpath="BAC_Data/Noise_sens"): 
    #with pre-estimator
    subs=os.listdir(bacpath)
    
    resR=pd.DataFrame(columns=['HY','RC','TSC',"HYJ","RCJ","TSCJ",'bHY','bRC',
        'bTSC',"bHYJ","bRCJ","bTSCJ",'oHY','oRC','oTSC',"oHYJ","oRCJ","oTSCJ",
        'hHY','hRC','hTSC',"hHYJ","hRCJ","hTSCJ",])
    fp = FloatProgress(min=0, max=len(subs))
    f2 = FloatProgress(min=0, max=1)
    display(fp)
    display(f2)
    for sbi,sb in enumerate(subs):
        fp.value=sbi
        if sb.find("TRUE")>=0 or not os.path.isdir(bacpath+"/"+sb):
            continue
        files=os.listdir(bacpath+"/"+sb)           
        sest=sb[sb.find("30")+2:sb.find("_")] 
        jns=np.float(sb[sb.find("noise")+5:sb.find("noise")+9])
        sest+="J" if sb.find("jump")>=0 else ""
        resR.loc[jns,sest]=0
        resR.loc[jns,"o"+sest]=0
        resR.loc[jns,"b"+sest]=0
        resR.loc[jns,"h"+sest]=0
        if True:
            qfiles=len(files)
            c=0
            for fi,f in enumerate(files,start=1):
                f2.value=fi/qfiles
                if f[-3:]!="npz":
                    qfiles-=1
                    continue
        
                with np.load(bacpath+"/"+sb+"/"+f) as arrs:
                    c_m0_NHY=arrs["npreest"]
                    c_m0_NBAC_HY_oracle_NR_Wk=arrs["NBAC_oracle"]
                    c_m0_NBAC_HY_VAB_NR_Wk=arrs["NBAC_VAB"]
                    c_m0_NBAC_HY_HY_NR_Wk=arrs["NBAC_HY"]
                    #print(arrs["NBAC_VAB"])    
                sbt=sb.replace("HY","TRUE").replace("RC","TRUE").replace("TSC","TRUE") 
                with np.load(bacpath+"/"+sbt+"/"+f) as arrs:
                    mt=arrs["highprec"] 
                norm=np.linalg.norm(mt)**2
                resR.loc[jns,sest]+=np.linalg.norm(mt-c_m0_NHY)**2/norm
                resR.loc[jns,"b"+sest]+=np.linalg.norm(mt-c_m0_NBAC_HY_VAB_NR_Wk)**2/norm
                resR.loc[jns,"h"+sest]+=np.linalg.norm(mt-c_m0_NBAC_HY_HY_NR_Wk)**2/norm
                resR.loc[jns,"o"+sest]+=np.linalg.norm(mt-c_m0_NBAC_HY_oracle_NR_Wk)**2/norm  
                c+=1
        resR.loc[jns,sest]/=c
        resR.loc[jns,"b"+sest]/=c
        resR.loc[jns,"h"+sest]/=c
        resR.loc[jns,"o"+sest]/=c  
    return resR

def noise_mag_pre_charts(res,savep):
    monochrome = (cycler('color', ['k']) * cycler('linestyle', ['-', '--', ':','-.']) )
    bcend=matplotlib.get_backend()
    matplotlib.use("pdf")
    for j in ["HY","RC","TSC"]:
        for i in range(2):
            fig,ax=plt.subplots()
            ax.set_prop_cycle(monochrome)
            suf="" if i==0 else "J"
            plt.plot(res.index,res[j+suf],label="pre-estimator")
            plt.plot(res.index,res["h"+j+suf],label="BAC with pair-wise beta")
            plt.plot(res.index,res["b"+j+suf],label="BAC with variance adjusted beta")
            plt.plot(res.index,res["o"+j+suf],label="BAC with oracle beta")
            
            #res[[j+suf,"o"+j+suf]].plot()
            #plt.ylabel("normalized MSE")
            plt.xlabel("$\kappa$")
            suf="No jumps :" if i==0 else "With jumps: "
            plt.title(suf+j)
            suf="nf" if i==0 else ""
            if j=="RC" and i==0:
                plt.legend()
            plt.savefig(savep+"/"+j+suf+".pdf")
    matplotlib.use(bcend)
    
    fig,ax=plt.subplots(2,3,figsize=(15,9))
    cnt=0
    fig.tight_layout(pad=3.0)
    for j in ["HY","RC","TSC"]:
        for i in range(2):
            ax[i][cnt].set_prop_cycle(monochrome)
            suf="" if i==0 else "J"
            ax[i][cnt].plot(res.index,res[j+suf],label="pre-estimator")
            ax[i][cnt].plot(res.index,res["h"+j+suf],label="BAC with pair-wise beta")
            ax[i][cnt].plot(res.index,res["b"+j+suf],label="BAC with variance adjusted beta")
            ax[i][cnt].plot(res.index,res["o"+j+suf],label="BAC with oracle beta")
            
            #res[[j+suf,"o"+j+suf]].plot()
            #plt.ylabel("normalized MSE")
            suf="No jumps :" if i==0 else "With jumps: "
            ax[i][cnt].set_title(suf+j)
            ax[i][cnt].set_xlabel("$\kappa$")
            suf="nf" if i==0 else ""
            if j=="RC" and i==0:
                ax[i][cnt].legend()
        cnt+=1   
     
```
